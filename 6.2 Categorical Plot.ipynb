```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) # Barplot
```

```python
sns.countplot(x='sex',data=tips) # Menghitung plot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow') # Boxplot
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h') # orient='h'-berdasarkan jam
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='Set1',hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='Set1',hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False) # jitter=False-meluruskan data
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1',dodge=True) # dodge=True-memisahkan data
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True) # dodge=true-memisahkan
```

```python
sns.violinplot(x='tip',y='day',data=tips,palette='rainbow')
sns.swarmplot(x='tip',y='day',data=tips,color='black',size=3) # isi diviolinplot
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar') # catplot-categorical plot
```
