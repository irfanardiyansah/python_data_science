```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips) # regregasi
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex') # memecah data
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex',palette='coolwarm') # memberi warna
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex',palette='coolwarm', markers=['o','v'], scatter_kws={'s':100}) # markers-merubah bentuk bulat, scatter_kws- besarnya size 100
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='sex',col='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',col='day')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',col='day',aspect=0.6,height=8) # aspect-lebar, heignt-tinggi
```
