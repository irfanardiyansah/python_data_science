```python
# Matplotlib - akan membuat sebuah diagram dgn sumbu x dan y

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,5,11) # linspace-membuat data array secara otomatis(dimulai 1, sampai 5, data ada 11)
```

```python
x
```

```python
y = x ** 2 # **(pangkat) x pangkat 2
```

```python
y
```

```python
plt.plot(x, y, 'r') # Membuat plot dgn data x, y, warna merah
plt.xlabel('X Axis') # Memberi nama pada sumbu X
plt.ylabel('Y Axis') # Memberi nama pada sumbu Y
plt.title('Data Contoh') # Memberi Judul plot
plt.show() # Hanya menampilkan plot tanpa informasi text dari x, y, title
```

```python
#subplot - membuat 1 kanvas dgn 2 plot

plt.subplot(1,2,1) # Membuat subplot1 pada satu kanvas (1 row, 2 kolom,1 plot)
plt.plot(x,y, 'r--')  # X dibawah, Y diatas, warna merah garis putus-putus
plt.subplot(1,2,2) # Membuat subplot2 pada kanvas (1 row, 2 kolom, 2 plot)
plt.plot(y, x, 'b*-') # Y dibawa, X diatas, warna biru dgn bintang garis
```

```python
# Custome ukuran plot - figure - tambahkan set untuk memberi nama pd label dan title

fig = plt.figure()
                                          # add_axes untuk memberi ukuran kanvas
axes = fig.add_axes([0.1, 0.1, 0.7, 1.5]) # 0.1 lebar Y ke X, 0.1 Tinggi X ke Y, 0.7 Lebar kanvas, 1.5 tinggi kanvas

axes.plot(x,y,'b')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Contoh Judl')
```

```python
fig = plt.figure()

a1 = fig.add_axes([0.1, 0.1, 0.9, 0.8])
a2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

a1.plot(x,y,'g')
a1.set_xlabel('X Besar')
a1.set_ylabel('Y_Besar')
a1.set_title("Plot Besar")

a2.plot(x,y,'b')
a2.set_xlabel('X Kecil')
a2.set_ylabel('Y Kecil')
a2.set_title('Plot Kecil')
```

```python
# untuk memperjelas ukuran kanvas dan mempertajam gambar
fig = plt.figure(figsize=(8,4), dpi=100) # figsize-ukuran kanvas yg akan dibuat l8,t4, dpi=100 resolusi
```

```python
fig, a1 = plt.subplots(figsize=(5,4), dpi=100)

a1.plot(x,y,'r')
a1.set_xlabel('X')
a1.set_ylabel('Y')
a1.set_title("Plot Besar")

fig.savefig("contoh.png",dpi=100) # Save dgn format png
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x ** 2, label="Data x pangkat 2")
axes.plot(x, x ** 3, label="Data x pangkat 3")


axes.legend(loc=2) # loc0,loc2-kiri atas, loc1-kanan atas, loc3- kiri bawah, loc4-kanan bawah
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color="red", linewidth=0.25)
axes.plot(x, x+2, color="red", linewidth=0.50)
axes.plot(x, x+3, color="red", linewidth=0.75)
axes.plot(x, x+4, color="red", linewidth=1)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color="green", lw=3, linestyle='-')
axes.plot(x, x+2, color="green", lw=3, ls='-.')
axes.plot(x, x+3, color="green", lw=3, ls=':')
axes.plot(x, x+4, color="green", lw=3, ls='--')
```
